### 독립과 배반은 다르다

- 독립 $P(A \cap B) = P(A)P(B)$
    - 사건 A의 발생 여부가 사건 B에 영향을 미치지 않음
- 배반 $P(A \cap B) = 0$
    - mutual exclusive 하다는 것.

### i.i.d

- independent and identically distributed
- 독립이고, $P(A \cap B) = P(A)P(B)$
    - 사건 A의 발생 여부가 사건 B에 영향을 미치지 않음
- 모든 확률 변수가 같은 확률 분포를 지님
- 많은 이론들이 iid라는 가정 하에 이뤄지고 있음. 애초에 저 룰을 따르지 않으면 적용할 수 없는 이론이 있음.

i.i.d. 가정은 많은 통계적 방법과 모델에서 기본적으로 사용됩니다. 예를 들어, 큰 수의 법칙과 중심 극한 정리는 i.i.d. 확률 변수들에 대한 가정 하에 성립합니다. 이러한 이론들은 표본 평균이 모집단의 평균으로 수렴하거나, 정규 분포에 수렴하는 등의 결과를 도출할 때 필요합니다.

i.i.d. 가정은 또한 기계 학습에서 데이터를 처리하거나, 실험 설계에서 표본을 추출하는 방식에도 영향을 미칩니다. 데이터 포인트들이 i.i.d.라는 가정 하에, 학습 알고리즘이나 통계적 추정이 효율적으로 작동할 수 있습니다. 데이터가 i.i.d. 가정을 만족하지 않을 경우, 예측 모델의 성능이 저하되거나 편향될 수 있습니다.

### 확률 분포

- 이산형 - discrete (pdf, 확률 질량 함수)
- 연속형 - continuous (pdf, 확률 밀도 함수)

### 표본 평균도 확률 변수다

- (표본이 무작위로 추출되기 때문에) 표본 평균도 확률변수이기에 평균과 분산이 존재한다.
    - 표본 평균의 평균 : 표본 평균의 기대값이 모집단 평균과 같다
        - $\mathbb{E}[\overline{X}] = \mu$
    - 표본 평균의 분산 : 표본 평균의 분산이 모집단 분산을 표본의 크기로 나눈 값과 같다
        - $\text{Var}(\overline{X}) = \frac{\sigma^2}{n}$
        - 즉, 표본 크기가 커질수록 표본 평균의 분산은 작아지며, 이는 표본 평균이 모집단 평균에 더 가까워진다는 것을 의미합니다.
    - 표본 평균의 평균과 분산은 `큰 수의 법칙`과 `중심극한정리(Central Limit Theorem, CLT)`의 기반을 이룹니다. 
        - 큰 수의 법칙에 따르면, 표본 크기가 충분히 크면 표본 평균이 모집단 평균에 수렴합니다. 
        - 중심극한정리는 표본 크기가 커질수록 표본 평균의 분포가 정규 분포에 점점 더 가까워진다는 것을 설명합니다.
